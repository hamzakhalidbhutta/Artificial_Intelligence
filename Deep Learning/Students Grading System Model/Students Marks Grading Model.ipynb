```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7])
df.head()
```

```python
df['obtained'] = df.s1 + df.s2 + df.s3 + df.s4 + df.s5
df.head()
```

```python
df['percentage'] = df.obtained / 500 * 100
df.head()
```

# Apply Symbolic AI, Classical programming, Traditional programming, Rule based programming 
 - Rules are always defined by developer
 - Rules + input = output

```python
def grade ( x ):
    if x >= 80:
        return "A+"
    elif x >= 70:
        return "A"
    elif x >= 60:
        return "B"
    elif x >= 50:
        return "C"
    elif x >= 40:
        return "D"
    elif x > 33:
        return "E"
    else:
        return "F"
```

```python
df["grade"] = df.percentage.apply(grade)
df.head()
# df.tail()
```

```python
df.grade.value_counts()
```

```python
dic = {
    "A+": 1,
    "A": 2,
    "B": 3,
    "C":4,
    "D": 5,
    "E":6,
    "F":7    }
df["grade1"] = df.grade.map(dic)
df.head()
```

# Finding grades using Deep Learning
 - input + output = rules

```python
df1 = df[["s1","s2","s3","s4","s5","grade"]]
df1.head(2)
df1 = df1.sample(frac=1)
df1.head()
```

```python
X = df1.loc[:,'s1':"s5"]
X.head()
```

```python
Y = df1[['grade']]
# Y = df1['grade']
Y.head()
```

# Vectorization to Y
 - One-Hot encoding

```python
Y = df1['grade']
Y = pd.get_dummies(Y)
Y.head()
```

```python
Y.drop_duplicates()
```

# Importing Libraries


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam,SGD
```

# Create model network structure
    - first-layer (input_shape = X.no_of_columns)
    - last-layers (output=Y.uique_labels(unique classes))
        - activation = according to the problem
    - Dense (output)

```python
model = Sequential()
```

```python
model.add(Dense(32, activation='relu', input_shape=(5,)))
# model.add(Dense(16, activation='relu', input_shape=(5,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(7, activation='softmax'))
```

```python
model.summary()
```

# selecting loss function based on the problem

```python
model.compile( optimizer=SGD(lr=0.002),
             loss='categorical_crossentropy',
             metrics=['acc'])
```

```python
model.fit(X,Y,epochs=30)
```
