```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(np.random.randint(10,32, size=(5000,5)),columns=["s1","s2","s3","s4","s5"])
df1.head()
```

```python
df2 = pd.DataFrame(np.random.randint(33,39, size=(5000,5)),columns=["s1","s2","s3","s4","s5"])
df2.head()
```

```python
df3 = pd.DataFrame(np.random.randint(40,49, size=(5000,5)),columns=["s1","s2","s3","s4","s5"])
df3.head()
```

```python
df4 = pd.DataFrame(np.random.randint(50,59, size=(5000,5)),columns=["s1","s2","s3","s4","s5"])
df4.head()
```

```python
df5 = pd.DataFrame(np.random.randint(60,69, size=(5000,5)),columns=["s1","s2","s3","s4","s5"])
df5.head()
```

```python
df6 = pd.DataFrame(np.random.randint(70,79, size=(5000,5)),columns=["s1","s2","s3","s4","s5"])
df6.head()
```

```python
df7 = pd.DataFrame(np.random.randint(80,89, size=(5000,5)),columns=["s1","s2","s3","s4","s5"])
df7.head()
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7])
df.head()
```

```python
df['obtained'] = df.s1 + df.s2 + df.s3 + df.s4 + df.s5
df.head()
```

```python
df['percentage'] = df.obtained / 500 * 100
df.head()
```

# Apply Symbolic AI, Classical programming, Traditional programming, Rule based programming 
 - Rules are always defined by developer
 - Rules + input = output

```python
def grade ( x ):
    if x >= 80:
        return "A+"
    elif x >= 70:
        return "A"
    elif x >= 60:
        return "B"
    elif x >= 50:
        return "C"
    elif x >= 40:
        return "D"
    elif x > 33:
        return "E"
    else:
        return "F"
```

```python
df["grade"] = df.percentage.apply(grade)
df.head()
# df.tail()
```

```python
df.grade.value_counts()
```

```python
dic = {
    "A+": 1,
    "A": 2,
    "B": 3,
    "C":4,
    "D": 5,
    "E":6,
    "F":7    }
df["grade1"] = df.grade.map(dic)
df.head()
```
